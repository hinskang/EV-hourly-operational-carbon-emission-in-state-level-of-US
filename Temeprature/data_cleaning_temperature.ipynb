```python
# 读入温度数据，数据格式为df，利用index_col=""设列GHCN_ID为行索引
f1path="./hly-temp-allall.csv"
df_temp=pd.read_csv(f1path,index_col="GHCN_ID")
df_temp
```

```python
#由于usecols没有切片功能，提取前六列所需数据
df_temp=df_temp.loc[:,"month":"comp_flag_HLY-TEMP-NORMAL"]
df_temp
```

```python
# 读入观测站inventory数据，利用usecols=[]读取指定的列，利用index_col=""设列ID为行索引
f2path="./hly_inventory.xlsx"
df_inventory=pd.read_excel(f2path,usecols=["ID","STATE"],index_col="ID")#只读取前两列数据，设列ID为行索引index
df_inventory
```

```python
# 利用df.drop(index=)函数，删除含del_ID的行的数据。默认axis=0,删除行的元素。axis=1,删除列的元素。
# 不能inplace=True，否则只会返回None
df_temp=df_temp.drop(index=del_ID)
```

# 计算各个州每小时的平均温度

```python
# 从全部温度数据筛选出各州的数据后，先利用.drop（df[df[""所在列]==删除值].index）剔除含有-9999，即缺失值的行
# 然后直接使用分类汇总函数（groupby）统计每小时的平均温度，无需先对各州数据按照时间降序排列数据才能做分类汇总
# 最后对分类汇总后的数据按时间降序排列，并重设行索引，并drop掉原行索引，保证数据列按时间降序排列，并且行索引是从0开始。
# 如果不重设行索引，对分类汇总后的数据按时间降序排列后，行索引是从一个大数至0递减的。
```

```python
# 删除缺失值的测试过程
# .drop后必须要赋值
# df_AL=df_AL.drop(df_AL[df_AL["HLY-TEMP-NORMAL"]==-9999].index)

# .rename的inplace不要等于True，否则会warning。如果inplace=True，也用等号赋值，就会返回None。
# df_AL.rename(columns={"HLY-TEMP-NORMAL":"HLY_TEMP_NORMAL"})

# any函数检查-9999值是否去除成功
# any(df_AL.HLY_TEMP_NORMAL == -9999)
```

```python
# 按多列对df数据进行排序

# 使用排序函数.sort_values()对df按多列的值进行排序,ascending=False降序排列。
# 语法：df.sort_values(by=["按哪些列排序，顺序表明优先级",""],ascending=False)[["需要显示的数据列","",""]]
# kind=mergesort，使用稳定的排序算法
# 例子:
# df_AL=df_AL.sort_values(by=["month","day","hour"],ascending=False)[["month","day","hour","HLY-TEMP-NORMAL","meas_flag_HLY-TEMP-NORMAL","comp_flag_HLY-TEMP-NORMAL"]]
```
